```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')  # to remove warning
```

```python
#loading the dataset
df=pd.read_csv('/Users/binoddhami/Desktop/Hotel_Booking/hotel_bookings 2.csv')
```

```python
##exploratory data analysis and data cleaning
df.head(10)
```

```python
#to show number of rows and columns
df.shape
```

```python
df.columns
```

```python
#to see null values
df.info()
```

```python
#change datatype
df['reservation_status_date']=pd.to_datetime(df['reservation_status_date'],format='%d/%m/%Y')
```

```python
df.info()
```

```python
#own
df['adr'] = df['adr'].astype(int)
```

```python
df.info()
```

```python
df['adr'] = df['adr'].astype(float)
```

```python
df.info()
```

```python
df.describe(include='object')
```

```python
#ddescring each rowes
for col in df.describe(include='object').columns:
    print(col)
    print(df[col].unique())
    print('-'*50)
```

```python
#identifying missing values
df.isnull().sum()
```

```python
#removing column their null values
df.drop(['company','agent'],axis=1,inplace=True)
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
#removing the outliers 
df=df[df['adr']<5000]
```

```python
#we have identyfy the dataset
#we have cleaned the dataset
# now analysis the data that we have cleaned
```

```python
#how much perc are cancelling the reservations
cancelled_perc=df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)
```

```python
plt.figure(figsize=(5,4))
plt.title('reservation status count')
plt.bar(['Not cancelled','Canceled'],df['is_canceled'].value_counts())
plt.show()
```

```python
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel',hue='is_canceled',data=df,palette='Blues')
plt.title('reservation status in different hotels',size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservation')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[[('adr')]].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[[('adr')]].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('avergae daily rate in city and resort hotel',fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```
